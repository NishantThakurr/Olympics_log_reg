## FILLING MISSING VALUES

```python

df = df.fillna({'Medal':'0'})
```

```python
df.head(10)
```

```python
df = df.fillna(df.median())
```

```python
df['Team'].value_counts()
```

```python
df = df.drop('Event',axis=1)
```

```python
df = df.drop('City',axis=1)
```

```python
df.info()
```

```python
df['Sex']=df['Sex'].map({'M':1,'F':0})
df['Season']=df['Season'].map({'Summer':1,'Winter':0})
```

```python

# pd.options.display.max_rows=None
df['Sport'].value_counts()
```

## DUMMIES

```python
df['Age']=df['Age'].astype(int)
df['Weight']=df['Weight'].astype(int)
df['Height']=df['Height'].astype(int)
```

```python
age_columns = pd.get_dummies(df['Age'],drop_first=True)
height_columns = pd.get_dummies(df['Height'],drop_first=True)
weight_columns = pd.get_dummies(df['Weight'],drop_first=True)
sport_columns = pd.get_dummies(df['Sport'],drop_first=True)
medal_columns = pd.get_dummies(df['Medal'],drop_first=True)
year_columns = pd.get_dummies(df['Year'],drop_first=True)
```

```python
year_columns

year1900_to_1924= year_columns.loc[:,1900:1924].max(axis=1)
year1928_to_1976= year_columns.loc[:,1928:1976].max(axis=1)
year1980_to_1998= year_columns.loc[:,1980:1998].max(axis=1)
year2000_to_2016= year_columns.loc[:,2000:2016].max(axis=1)
year_columns=pd.concat([year1900_to_1924,year1928_to_1976,year1980_to_1998,year2000_to_2016],axis=1)
year_columns=year_columns.fillna(0)
year_columns.columns.values
year_columns.columns=['year1900_to_1924','year1928_to_1976','year1980_to_1998','year2000_to_2016']


```

```python
year_columns
```

```python

age11_to_19= age_columns.loc[:,11:19].max(axis=1)
age20_to_25= age_columns.loc[:,20:25].max(axis=1)
age26_to_50= age_columns.loc[:,26:50].max(axis=1)
age51_to_97= age_columns.loc[:,51:97].max(axis=1)

age_columns=pd.concat([age11_to_19,age20_to_25,age26_to_50,age51_to_97],axis=1)
age_columns=age_columns.fillna(0)
age_columns.columns.values
age_columns.columns=['age11_to_19','age20_to_25','age26_to_50','age51_to_97']

```

```python
age_columns
```

```python
pd.options.display.max_columns=None

height128_to_150 = height_columns.loc[:,128:150].max(axis=1)
height151_to_170 = height_columns.loc[:,151:170].max(axis=1)
height171_to_190 = height_columns.loc[:,171:190].max(axis=1)
height191_to_226 = height_columns.loc[:,191:226].max(axis=1)

height_columns=pd.concat([height128_to_150,height151_to_170,height151_to_170,height191_to_226],axis=1)
height_columns=height_columns.fillna(0)
height_columns.columns.values
height_columns.columns=['height128_to_150','height151_to_170','height151_to_170','height191_to_226']

# height_columns
```

```python
height_columns
```

```python
# weight_columns
weight28_to_50 = weight_columns.loc[:,28:50].max(axis=1)
weight51_to_80= weight_columns.loc[:,51:80].max(axis=1)
weight81_to_150 = weight_columns.loc[:,81:150].max(axis=1)
weight150_to_214 = weight_columns.loc[:,150:214].max(axis=1)

weight_columns=pd.concat([weight28_to_50,weight51_to_80,weight81_to_150,weight150_to_214],axis=1)
weight_columns=weight_columns.fillna(0)
weight_columns.columns.values
weight_columns.columns=['weight28_to_50','weight51_to_80','weight81_to_150','weight150_to_214']
```
